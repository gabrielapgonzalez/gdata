```python
bureau_pd = dic["bureau"].compute()
```

```python
n_prev_app = pd.DataFrame(data = bureau_pd.SK_ID_CURR.value_counts())
n_prev_app.columns = ["N_PREV_APP"]
```

```python
df = application_train_pd.set_index("SK_ID_CURR").join(n_prev_app)
```

```python
print("Porcentagem de NaN train")
print(df.N_PREV_APP.isnull().sum()*100/df.shape[0])
```

```python
dft = application_test_pd.set_index("SK_ID_CURR").join(n_prev_app)
```

```python
print("Porcentagem de NaN test")
print(dft.N_PREV_APP.isnull().sum()*100/df.shape[0])
```
